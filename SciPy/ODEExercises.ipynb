(a) Use Euler's method (not odeint) to solve the following first-order differential equation with given initial condition y(0) = 0.

dy/dx=x+2y

Find the value of y when x = 1, y(1), using two different step sizes: 0.25 then 0.02.

```python
import numpy as np

def Euler(xmax,dx):
    a = -9.81 #g in m/s^2
    nsteps = int(xmax/dx)
    x = np.zeros(nsteps)
    dydx = np.zeros(nsteps)
    y = np.zeros(nsteps)
 
    # set initial conditions, redundant in this example
    x[0] = 0.
    y[0] = 0.

    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        dydx[i+1] = x[i] + 2*y[i]
        y[i+1] = y[i]+ dydx[i] * dx
        
    return x,y

x1,y1 = Euler(100,0.25)
x2,y2 = Euler(100,0.02)
ii = (x1==1)
jj = (x1==1)
print y1[ii]
print y2[jj]
```

```python
def Projectile_deriv(y,t,g,vterm):
    ay = g #g in m/s^2
    vx0 = y[1]
    vy0 = y[3]
    
    dvxdt = (ay*vx0/vterm**2) * np.sqrt(vy0**2 + vx0**2)
    dvydt = ay * (1 + (vy0/vterm**2) * np.sqrt(vy0**2 + vx0**2))
    
    return np.array([vx0,dvxdt,vy0,dvydt])
```

(a) Write a function and interact widget that solves the ODE for the nonlinear pendulum with damping and driving. The parameters, ℓ, q, fd, and Ωd should be passed into the interact as sliders. Use initial values of θ0 =π/2, ω0=0 and a time span of 50 seconds.

Check that the solution looks as it did before for the case: ℓ=1, q=0, fd=0, Ωd=0. Play with the sliders to see how the solution depends on the parameters.

```python
def deriv(current,t,ell,q,fd,Omega):
    
    g = 9.81
    omega0 = current[0]
    theta0 = current[1]

    dtheta = omega0
    domega = -(g/ell)*np.sin(theta0) - q*omega0 - fd*np.sin(Omega*t)

    return np.array([domega,dtheta,omega0,theta0])

def Pendulum(omega = 0, th0 = np.pi/2, time=50, ell = 1.0,\
               q = 0.0, fd = 0.0, Omega = 0.0):

    t = np.arange(0.,time+0.001,0.001)

    # initial conditions
    init = np.array([omega0, th0])

    # solve
    result = odeint(deriv, init ,t, args=(ell,q,fd,Omega))

    omega = result[:,0]
    theta = result[:,1]
    
    # plot relationships
    
    ####### PLOT E/m vs. t #######
    E = g * ell * (1-np.cos(result[:,1])) + 0.5 * (ell**2 * result[:,0]**2)

    plt.plot(t,E,label="W/ Damping = " + str(q) + " & Driving Force = " + str(fd))
    plt.xlabel("time (sec)", fontsize=15)
    plt.ylabel(r'$E/m (Joules/kg)$', fontsize=20)
    plt.legend()
    plt.axis([0,50,0,20])
    plt.show()

    plt.plot(t,theta,label="W/ Damping = " + str(q) + " & Driving Force = " + str(fd))
    plt.xlabel("time (sec)", fontsize=15)
    plt.ylabel(r'$\theta (rad)$', fontsize=20)
    plt.legend()
    plt.axis([0,50,-3,3])
    plt.show()
    
    plt.plot(t,omega,label="W/ Damping = " + str(q) + " & Driving Force = " + str(fd))
    plt.xlabel("time (sec)", fontsize=15)
    plt.ylabel(r'$\omega (rad/sec)$', fontsize=20)
    plt.legend()
    plt.axis([0,50,-10,10])
    plt.show()
    
    plt.plot(theta,omega,label="W/ Damping = " + str(q) + " & Driving Force = " + str(fd))
    plt.xlabel(r'$\theta (rad/sec)$', fontsize=20)
    plt.ylabel(r'$\omega (rad/sec)$', fontsize=20)
    plt.legend()
    plt.axis([0,4,-10,10])
    plt.show()

Pendulum()
v = interact(Pendulum,ell=(0.,15.),q=(0.,10.),fd=(0.,10.),Omega=(0.,10.), full_output=1)
```
