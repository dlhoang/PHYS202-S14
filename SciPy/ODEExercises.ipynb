(a) Use Euler's method (not odeint) to solve the following first-order differential equation with given initial condition y(0) = 0.

dy/dx=x+2y

Find the value of y when x = 1, y(1), using two different step sizes: 0.25 then 0.02.

```python
import numpy as np

def Euler(xmax,dx):
    a = -9.81 #g in m/s^2
    nsteps = int(xmax/dx)
    x = np.zeros(nsteps)
    dydx = np.zeros(nsteps)
    y = np.zeros(nsteps)
 
    # set initial conditions, redundant in this example
    x[0] = 0.
    y[0] = 0.

    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        dydx[i+1] = x[i] + 2*y[i]
        y[i+1] = y[i]+ dydx[i] * dx
        
    return x,y

x1,y1 = Euler(100,0.25)
x2,y2 = Euler(100,0.02)
ii = (x1==1)
jj = (x1==1)
print y1[ii]
print y2[jj]
```

```python
import numpy as np

def EulerCromer(xmax,dx):
    a = -9.81 #g in m/s^2
    nsteps = int(xmax/dx)
    x = np.zeros(nsteps)
    dydx = np.zeros(nsteps)
    y = np.zeros(nsteps)
 
    # set initial conditions, redundant in this example
    x[0] = 0.
    y[0] = 0.

    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        dydx[i+1] = x[i] + 2*y[i]
        y[i+1] = y[i]+ dydx[i+1] * dx
        
    return x,y

x1,y1 = EulerCromer(100,0.25)
x2,y2 = EulerCromer(100,0.02)

def MidPoint(xmax,dx):
    a = -9.81 #g in m/s^2
    nsteps = int(xmax/dx)
    x = np.zeros(nsteps)
    dydx = np.zeros(nsteps)
    y = np.zeros(nsteps)
 
    # set initial conditions, redundant in this example
    x[0] = 0.
    y[0] = 0.

    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        dydx[i+1] = x[i] + 2*y[i]
        y[i+1] = y[i]+ 0.5 * (dydx[i] + dydx[i+1])* dx
        
    return x,y

x3,y3 = MidPoint(100,0.25)
x4,y4 = MidPoint(100,0.02)
```
