### Question 2

(a) Use Euler's method (not odeint) to solve the following first-order differential equation with given initial condition y(0) = 0.

dy/dx=x+2y

Find the value of y when x = 1, y(1), using two different step sizes: 0.25 then 0.02.

```python
import numpy as np

def Euler(xmax,dx):
    a = -9.81 #g in m/s^2
    nsteps = int(xmax/dx)
    x = np.zeros(nsteps)
    y = np.zeros(nsteps)
 
    # set initial conditions, redundant in this example
    x[0] = 0.
    y[0] = 0.

    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        y[i+1] = y[i]+ ((x[i] + 2*y[i]) * dx)
        if x[i] == 1.:
            print y[i]
        
    return x,y

x1,y1 = Euler(1.25,0.25)
x2,y2 = Euler(1.2,0.02)
```

(b) Plot the results from part (a) for the two different step sizes with the exact solution

y=0.25e2x−0.5x−0.25

on one plot. Include a legend. Which step size gives a more accurate result?

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,1,100)
y = 0.25 * np.exp(2*x) - 0.5*x - 0.25

plt.title("y vs x")
plt.plot(x,y,'g-',label="exact")
plt.plot(x1,y1, 'r-', label="step size = 0.25")
plt.plot(x2,y2, 'b-', label="step size = 0.02")

plt.xlabel(r'$x$',fontsize=20)
plt.ylabel(r'$y$',fontsize=20)
plt.legend()
plt.show()
```

(c) Now solve the same problem using Euler-Cromer and the Midpoint method for the same pair of step sizes and plot the results along with the exact solution. Then also plot the difference between the approximate numerical solutions and the exact solution. Which combination of method and step-size gives the best result?

```python
import numpy as np

def EulerCromer(xmax,dx):
    a = -9.81 #g in m/s^2
    nsteps = int(xmax/dx)
    x = np.zeros(nsteps)
    y = np.zeros(nsteps)
 
    # set initial conditions, redundant in this example
    x[0] = 0.
    y[0] = 0.

    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        y[i+1] = y[i]+ ((x[i+1] + 2*y[i]) * dx)
        if x[i] == 1.:
            print y[i]
        
    return x,y

x1,y1 = EulerCromer(1.25,0.25)
x2,y2 = EulerCromer(1.2,0.02)

def MidPoint(xmax,dx):
    a = -9.81 #g in m/s^2
    nsteps = int(xmax/dx)
    x = np.zeros(nsteps)
    y = np.zeros(nsteps)
 
    # set initial conditions, redundant in this example
    x[0] = 0.
    y[0] = 0.

    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        y[i+1] = y[i]+ 0.5*((x[i] + 2*y[i] + x[i+1] + 2*y[i]) * dx)
        if x[i] == 1.:
            print y[i]
        
    return x,y

x3,y3 = MidPoint(1.25,0.25)
x4,y4 = MidPoint(1.2,0.02)

# actual result
x = np.linspace(0,1,5)
y = 0.25 * np.exp(2*x) - 0.5*x - 0.25

plt.title("y vs x")
plt.plot(x,y,'g-',label="exact")
plt.plot(x1,y1, 'r-', label="Euler-Cromer 0.25")
plt.plot(x2,y2, 'b-', label="Euler-Cromer 0.02")
plt.plot(x3,y3, 'y-', label="Midpoint 0.25")
plt.plot(x4,y4, 'b--', label="Midpoint = 0.02")

plt.xlabel(r'$x$',fontsize=20)
plt.ylabel(r'$y$',fontsize=20)
plt.legend(loc="upper left")
plt.show()
```
