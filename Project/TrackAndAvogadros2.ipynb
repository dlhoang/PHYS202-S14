```python
import BlobFinder2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import requests
from StringIO import StringIO
import os

def images(folder):
    """ returns a list of pictures in a folder to be looked at """
    pictures = []

    for file in os.listdir(folder):
        if file.endswith(".jpg"):
            filename = folder + "/" + file
            pictures.append(filename)

    return pictures

def movement(P,pictures,tau,delta):
    """ returns a list of all distances between blobs from one frame to another """
    distance = []
    
    # loop through all images
    for i in range(0,len(pictures)-1):
        # open one picture and get all blobs with luminance threshold tau
        picture1 = Image.open(pictures[i])
        frame1 = BlobFinder2.getBeads(P,picture1,tau)
        # open next picture and get all blobs with luminance threshold tau
        picture2 = Image.open(pictures[i+1])
        frame2 = BlobFinder2.getBeads(P,picture2,tau)
        # keep track of how many blobs in each picture
        # beads1 = BlobFinder.countBeads(P,picture1,tau)
        # beads2 = BlobFinder.countBeads(P,picture2,tau)
        
        # loop through every blob in first picture
        for blob in frame1:
            dist = []
            
            for j in xrange(len(frame2)-1):
                # have a list of all possibilities of distances between current blob with all blobs
                # in second frame and get the minimum
                dist.append(blob.distanceTo(frame2[j]))
                low = min(x for x in dist if x != 0.0)
                
            # if distance between two blobs is smallest and less than delta, add to list of distances
            if low < delta:
                distance.append(low)

    return distance

def writeFile(lst):
    """ write all distances found to a txt file """
    # turn list of distances into array in order to use savetxt
    myarray = np.asarray(distance)
    arr = myarray.reshape(len(distance), 1)
    np.savetxt('DistanceData2.txt', arr, fmt='%5.4e', header="Distance Data")

    out = open('DistanceData2.txt')
    y = out.read()
    print y
    out.close()
    
pictures = images("run4")
distance = movement(35.,pictures,210.,10.)
writeFile(distance)
```

- The self-diffusion constant D characterizes the stochastic movement of a molecule (bead) through a homogeneous medium (the water molecules) as a result of random thermal energy. The Einstein-Smoluchowski equation states that the random displacement of a bead in one dimension has a Gaussian distribution with mean zero and variance σ$^2$=2DΔt, where Δt is the time interval between position measurements. That is, a molecule's mean displacement is zero and its mean square displacement is proportional to the elapsed time between measurements, with the constant of proportionality 2D. For the test data, Δt = 0.5 so this is an estimate for D as well.
- The Stokes-Einstein relation asserts that the self-diffusion constant of a spherical particle immersed in a fluid is given by D=kT/6πηρ where, for the test data,

T = absolute temperature = 297 degrees Kelvin (room temperature)

η = viscosity of water = 9.135$10^{-4}$ N*s/m2 (at room temperature)

ρ = radius of bead = 0.5*10$^{-6}$ meters

and k is the Boltzmann constant. All parameters are given in SI units. The Boltzmann constant is a fundamental physical constant that relates the average kinetic energy of a molecule to its temperature. We estimate k by measuring all of the parameters in the Stokes-Einstein equation, and solving for k.

- Avogadro's number NA is defined to be the number of particles in a mole. By definition, k=R/NA, where the universal gas constant R is approximately 8.31457 J/K⋅mol. Use R/k as an estimate of Avogadro's number.

```python
def AvogadrosNumber(lst):
    """ returns estimate of Avogadro's number """
    delta_t = 0.5 # seconds
    temp_sigma = 0.0
    # loop through all distances
    for i in range(0,len(distance)):
        # equation to sigma squared
        temp_sigma = temp_sigma + (distance[i]* 0.175*10**(-6))**2
    sigma_squared = ("%.4e" % (temp_sigma / (2*len(distance))))
    # print out sigma squared
    
    print "sigma_squared " + str(sigma_squared)
    
    # equation of D -- in this case, D = sigma squared
    D = float(sigma_squared) /(2*delta_t)

    # estimating Boltzmann constant
    T = 297.0 # room temperature, Kelvin
    nu = 9.135*(10**(-4)) # viscosity of water, N*s/m^2
    ro = 0.5*(10**(-6)) # radius of beads, m
    k = ("%.4e" % ((D*6*np.pi*nu*ro) / T)) # Boltzmann constant, m^2*kg/s*K
    
    print "Boltzmann: " + str(k)

    """ estimating Avogadro's number """
    R = 8.31457 # J/K⋅mol
    A = ("%.4e" % (R/float(k)))
    err = ("%.4e" % (((float(A)-6.0221413e23) / 6.0221413e23) * 100))

    return A,err

Na,error = AvogadrosNumber(distance)
print "Avogadro's Number: " + str(Na)
print "Percent error: " + str(error)
```
